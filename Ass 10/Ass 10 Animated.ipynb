```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from celluloid import Camera
```

```python
# Function definition
def f(x_, t):
    return 0.5 * (x_ + t)**2 + 2 * np.sin(10 * (x_ - t))
```

```python
# Task 1
fig, ax = plt.subplots()
x = np.linspace(0, 10, 1000)
line, = ax.plot(x, f(x, 0))


def update(t):
    ax.clear() 
    ax.plot(x, f(x, t))  
    ax.set_xlabel('x')
    ax.set_ylabel('f(x, t)')
    ax.set_title(f'f(x, t) at t = {t:.2f}')

# Creating an animation
ani = FuncAnimation(fig, update, frames=np.linspace(0, 10, 200), repeat=False)
# Show the animation
plt.show()
```

# Task 2

```python
# Task 2
fig, ax = plt.subplots()
camera = Camera(fig)
x = np.linspace(0, 20, 1000)
# Iterating over time values
for t in np.linspace(0, 20, 200):
    ax.plot(x, f(x, t))
    ax.set_xlabel('x')
    ax.set_ylabel('f(x, t)')
    ax.set_title(f'f(x, t) at t = {t:.2f}')
    camera.snap()
# animation
animation = camera.animate()


plt.show()
```

```python
# Task 3
fig, ax = plt.subplots()
camera = Camera(fig)
x = np.linspace(0, 20, 1000)
for t in np.linspace(0, 20, 200):
    ax.plot(x, f(x, t))
    ax.set_xlabel('x')
    ax.set_ylabel('f(x, t)')
    ax.set_title(f'f(x, t) at t = {t:.2f}')


    ax.text(8, 8, f'Time: {t:.2f}', fontsize=12, color='blue')

    camera.snap()

animation = camera.animate()

plt.show()
```
