```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import spacy

# Load spaCy English model
nlp = spacy.load("en_core_web_sm")

# Custom tokenizer using spaCy for lemmatization
def spacy_tokenizer(text):
    tokens = nlp(text)
    return [token.lemma_ for token in tokens if not token.is_punct and not token.is_space]

# Define the text corpus
c = {
    'Lincoln1865': 'With malice toward none, with charity for all ...' +
                   'let us strive on to finish the work we are in ... ' +
                   'to do all which may achieve and cherish a just and lasting peace, ' +
                   'among ourselves, and with all nations.',
    'TrumpMay26': 'There is NO WAY (ZERO!) that Mail-In Ballots ' +
                  'will be anything less than substantially fraudulent.',
    'Wikipedia': 'In 1998, Oregon became the first state in the US ' +
                 'to conduct all voting exclusively by mail.',
    'FortuneMay26': 'Over the last two decades, about 0.00006% of total ' +
                    'vote-by-mail votes cast were fraudulent.',
    'TheHillApr07': 'Trump voted by mail in the Florida primary.',
    'KingJamesBible': 'Wherefore laying aside all malice, and all guile, and ' +
                      'hypocrisies, and envies, and all evil speakings',
}

# Create an instance of CountVectorizer with custom tokenizer
vectorizer = CountVectorizer(tokenizer=spacy_tokenizer, stop_words='english')

# Fit and transform the text data
term_document_matrix = vectorizer.fit_transform(c.values())

# Display the term-document matrix as a DataFrame
df = pd.DataFrame(term_document_matrix.toarray(), columns=vectorizer.get_feature_names_out(), index=c.keys())
print(df)
```

```python
# TASK 4
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def cosine_similarity_between_words(word1, word2, corpus):
    vectorizer = CountVectorizer(stop_words='english')
    X = vectorizer.fit_transform(corpus)
    
    word1_vector = X[:, vectorizer.vocabulary_[word1]].toarray().flatten()
    word2_vector = X[:, vectorizer.vocabulary_[word2]].toarray().flatten()
    
    # Compute cosine similarity
    similarity = cosine_similarity(word1_vector.reshape(1, -1), word2_vector.reshape(1, -1))[0, 0]
    
    return similarity

# Define the text corpus
corpus = [
    'With malice toward none, with charity for all ... let us strive on to finish the work we are in ... to do all which may achieve and cherish a just and lasting peace, among ourselves, and with all nations.',
    'There is NO WAY (ZERO!) that Mail-In Ballots will be anything less than substantially fraudulent.',
    'In 1998, Oregon became the first state in the US to conduct all voting exclusively by mail.',
    'Over the last two decades, about 0.00006% of total vote-by-mail votes cast were fraudulent.',
    'Trump voted by mail in the Florida primary.',
    'Wherefore laying aside all malice, and all guile, and hypocrisies, and envies, and all evil speakings'
]

# Compute cosine similarity between "malice" and "vote"
similarity_malice_vote = cosine_similarity_between_words('malice', 'vote', corpus)
print(f"Cosine Similarity between 'malice' and 'vote': {similarity_malice_vote:.4f}")

# Compute cosine similarity between "mail" and "vote"
similarity_mail_vote = cosine_similarity_between_words('mail', 'vote', corpus)
print(f"Cosine Similarity between 'mail' and 'vote': {similarity_mail_vote:.4f}")
```

```python
# TASK 6

import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Define the text corpus
corpus = [
    'With malice toward none, with charity for all ... let us strive on to finish the work we are in ... to do all which may achieve and cherish a just and lasting peace, among ourselves, and with all nations.',
    'There is NO WAY (ZERO!) that Mail-In Ballots will be anything less than substantially fraudulent.',
    'In 1998, Oregon became the first state in the US to conduct all voting exclusively by mail.',
    'Over the last two decades, about 0.00006% of total vote-by-mail votes cast were fraudulent.',
    'Trump voted by mail in the Florida primary.',
    'Wherefore laying aside all malice, and all guile, and hypocrisies, and envies, and all evil speakings'
]

# Create an instance of TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words='english')

# Fit and transform the text data to obtain the TF-IDF matrix
tfidf_matrix = vectorizer.fit_transform(corpus)

# Function to compute cosine similarity between words using TF-IDF matrix
def cosine_similarity_tfidf(word1, word2, tfidf_matrix, vectorizer):
    word1_vector = tfidf_matrix[:, vectorizer.vocabulary_[word1]].toarray().flatten()
    word2_vector = tfidf_matrix[:, vectorizer.vocabulary_[word2]].toarray().flatten()
    
    # Compute cosine similarity
    similarity = cosine_similarity(word1_vector.reshape(1, -1), word2_vector.reshape(1, -1))[0, 0]
    
    return similarity

# Compute cosine similarity between "malice" and "vote" using TF-IDF
similarity_malice_vote_tfidf = cosine_similarity_tfidf('malice', 'vote', tfidf_matrix, vectorizer)
print(f"Cosine Similarity between 'malice' and 'vote' (TF-IDF): {similarity_malice_vote_tfidf:.4f}")

# Compute cosine similarity between "mail" and "vote" using TF-IDF
similarity_mail_vote_tfidf = cosine_similarity_tfidf('mail', 'vote', tfidf_matrix, vectorizer)
print(f"Cosine Similarity between 'mail' and 'vote' (TF-IDF): {similarity_mail_vote_tfidf:.4f}")
```
