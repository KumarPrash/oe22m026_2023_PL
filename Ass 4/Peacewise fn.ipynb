# Passing function Arguments.

```python
def twice(w):
    w*=2
    return w
    

```

```python
v=[2,5,1]
print("New value",twice(v))
print("Original value",v)
## This happens since python pass the object by reference as an argument. It modifies the list in place
```

```python
import copy
def twice_(w):
    lst_= copy.deepcopy(w)
    lst_=lst_ *2
    return lst_
    
```

```python
v=[2,5,1]
print(twice_(v))
```

```python
v="Prashant"
print(twice_(v))
```

```python
# In both case same thing is happening.
```

# Piecewise functions

```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
def f(x):
    return x*(np.sin(x**2))
```

```python
x=np.linspace(-5,5,1000001) # As mentioned in question
y=f(x)
# Plot the x*sin(x^2) function
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of x*sin(x^2)')

plt.show()
```

```python
# For cosx graph.
def f(x):
    return np.cos(x)
x=np.linspace(-5,5,1000001)
y=f(x)
plt.plot(x,y)
plt.title("Plot of cosx")
```
