# Predator-prey model

```python
#TASK1 : Given initial conditions $ r(0) = 5$ and $ x(0) = 2, $ solve for r(t) and x(t) and plot the solution for 0 ≤ t ≤ 70.
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters
alpha = 1.1
beta = 0.4
delta = 0.1
gamma = 0.1

# Define the initial conditions
r0 = 5
x0 = 2

# Define the time span
t = np.linspace(0, 70, 1000)  # Create a time array

# Create empty arrays to store the solutions
r = np.zeros(len(t))
x = np.zeros(len(t))

# Set the initial conditions
r[0] = r0
x[0] = x0

# Use Euler's method to solve the ODEs
for i in range(1, len(t)):
    drdt = alpha * r[i-1] - beta * r[i-1] * x[i-1]
    dxdt = delta * r[i-1] * x[i-1] - gamma * x[i-1]
    r[i] = r[i-1] + drdt * (t[i] - t[i-1])
    x[i] = x[i-1] + dxdt * (t[i] - t[i-1])

# Plot the solutions
plt.figure(figsize=(10, 6))
plt.plot(t, r, label='Rabbits', color='blue')
plt.plot(t, x, label='Foxes', color='red')
plt.xlabel('Time (t)')
plt.ylabel('Population')
plt.title('Rabbit and Fox Population Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#TASK2: The phase plot of the solution consists of points (x(t), r(t)) for various t values.
#Prepare a fgure (phase portrait) with phase plots of, say, 10 solutions, one each for randomly
#chosen initial values r(0) and x(0) between 1 and 9.

import numpy as np
import matplotlib.pyplot as plt

# Define the parameters
alpha = 1.1
beta = 0.4
delta = 0.1
gamma = 0.1

# Define the time span
t = np.linspace(0, 70, 1000)

# Create an array of 10 random initial conditions
np.random.seed(0)  # Set a seed for reproducibility
initial_conditions = np.random.uniform(1, 9, size=(10, 2))

# Create a figure for the phase portrait
plt.figure(figsize=(10, 8))
plt.title('Phase Portrait of Rabbit and Fox Populations')

# Plot phase plots for the 10 solutions
for i in range(10):
    r0, x0 = initial_conditions[i]
    r = np.zeros(len(t))
    x = np.zeros(len(t))
    r[0] = r0
    x[0] = x0

    for j in range(1, len(t)):
        drdt = alpha * r[j-1] - beta * r[j-1] * x[j-1]
        dxdt = delta * r[j-1] * x[j-1] - gamma * x[j-1]
        r[j] = r[j-1] + drdt * (t[j] - t[j-1])
        x[j] = x[j-1] + dxdt * (t[j] - t[j-1])

    plt.plot(x, r, label=f'r(0)={r0}, x(0)={x0}')

# Add labels and legend
plt.xlabel('Fox Population (x)')
plt.ylabel('Rabbit Population (r)')
plt.legend()
plt.grid(True)
plt.show()
```
